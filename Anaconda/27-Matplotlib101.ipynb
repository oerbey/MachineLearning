```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# for jupyter notebook
```

```python
ageList = [10,20,30,30,40,50,60,70,78,80]
kgList = [20,50,80,85,86,87,70,90,95,90]
```

```python
numpyAgeList = np.array(ageList)
numpyKgList = np.array(kgList)
```

```python
plt.plot(numpyAgeList,numpyKgList,"b") # g is green for line
plt.xlabel("Age")
plt.ylabel("kg")
plt.title("Age - kg graph")
plt.show() # to show in other script editors
```

```python
numpyArray1 = np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1 ** 3
```

```python
numpyArray2
```

```python
plt.plot(numpyArray1, numpyArray2, "g*-") # "b--"
```

```python
plt.subplot(1,2,1) # how many rows?, how many graphics, draw 1st
plt.plot(numpyArray1,numpyArray2,"r--")
plt.subplot(1,2,2) # 1 row, 2 column, 2nd graph
plt.plot(numpyArray2, numpyArray1,"b*-")
```

```python
myFigure = plt.figure() # it's like empty canvas
```

```python
myFigure = plt.figure() # it's like empty canvas
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9]) #width, height, width, height
figureAxes.plot(numpyArray1,numpyArray2,"g")
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("Graphic Header ")
```

```python
figure2 = plt.figure()

axis1 = figure2.add_axes([0.1,0.1,0.7,0.7]) #x, y, width, height
axis2 = figure2.add_axes([0.2,0.4,0.3,0.3]) 

axis1.plot(numpyArray1,numpyArray2,"g")
axis1.set_xlabel("Axis X")
axis1.set_ylabel("Axis Y")
axis1.set_title("Header")

axis2.plot(numpyArray2,numpyArray1,"b")
axis2.set_xlabel("Axis X")
axis2.set_ylabel("Axis Y")
axis2.set_title("Small Header")
```

```python
(myFig, myAxis) = plt.subplots(nrows = 1, ncols = 2) # it returns tuple 
#myAxis.plot(numpyArray1,numpyArray2,"r")
type(myAxis)

for axis in myAxis:
    axis.plot(numpyArray1,numpyArray2,"g")
    axis.set_xlabel("Axis X")
    
plt.tight_layout() # keep distance between graphs 
    
#myFigure to show in other code editors
```

```python
myAxis.plot(numpyArray1,numpyArray2,"r")
```

```python
#newFig = plt.figure(figsize=(6,6)) #with and height is tuple if you don't enter, it will determine by axis
# can set dpi as well
newFig = plt.figure(dpi=150)

newAxis = newFig.add_axes([0.1,0.1,0.9,0.9])

newAxis.plot(numpyArray1,numpyArray2,label = "Numpy Array ** 3")
newAxis.plot(numpyArray1, numpyArray1 ** 2, label = "Numpay Array ** 2")
newAxis.legend(loc=6) # to return label and choose location
```
